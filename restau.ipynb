```python
total_rev = lis_for_rev[6]
print(f'1st is : {total_rev}')
total_rev = re.sub(r'[^a-zA-Z0-9\']+', ' ',total_rev )
print(f'2nd is : {total_rev}')
total_rev = re.sub("\W\'", " ", total_rev)
print(f'3rd is : {total_rev}')
total_rev=re.sub('^\s+',"",total_rev)
print(f'4th is : {total_rev}') 
total_rev=re.sub('\s+$',"",total_rev)
print(f'5th is : {total_rev}')
total_rev = total_rev.lower()
print(f'6th is : {total_rev}')
```

```python
rev_result = total_rev.split()
rev_result
```

```python
lower_rev = []
for word in rev_result:
    if word.lower() not in stop_words:
        lower_rev.append(word.lower())
```

```python
from nltk.stem import WordNetLemmatizer
le = WordNetLemmatizer()
lem_word = []
for word in lower_rev:
    lem_word.append(le.lemmatize(word))
```

```python
t_rev = []
for i in range(len(rev_arr)):
    total_rev = lis_for_rev[i]
    total_rev = re.sub(r'[^a-zA-Z0-9\']+', ' ',total_rev )
    total_rev = re.sub("\W\'", " ", total_rev)
    total_rev=re.sub('^\s+',"",total_rev)
    total_rev=re.sub('\s+$',"",total_rev)
    total_rev = total_rev.lower()
    total_rev = total_rev.split()
    ps = PorterStemmer()
    total_rev = [ps.stem(word) for word in total_rev if not word in set(stopwords.words('english'))]
    total_rev = ' '.join(total_rev)
    t_rev.append(total_rev)
```

```python
total_rev
```

```python
t_rev
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 1500)
x = cv.fit_transform(t_rev).toarray()
y = dataset.iloc[:, 1].values
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state= 0)
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
shape = LogisticRegression()
shape.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = shape.predict(x_test)
print(f'The acc is : {accuracy_score(y_test, y_pred)*100}%')
```

```python
from sklearn.svm import SVC
classi = SVC()
classi.fit(x_train, y_train)
```

```python
y_pred = classi.predict(x_test)
print(f'The acc is : {accuracy_score(y_test, y_pred)*100}%')
```

```python
from sklearn.neighbors import KNeighborsClassifier
k_class = KNeighborsClassifier(n_neighbors=5)
k_class.fit(x_train, y_train)
```

```python
y_pred = k_class.predict(x_test)
print(f'The acc is : {accuracy_score(y_test, y_pred)*100}%')
# very bad mod
```
